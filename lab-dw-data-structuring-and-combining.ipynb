# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
# Your code goes here
marketing_customers_df = pd.read_csv('./datasets/marketing_customer_analysis_clean.csv')
display(marketing_customers_df)

## Cleaning ##
print(list(marketing_customers_df.columns)) # Review columns
# Apply column cleaning and selective rename
marketing_customers_df = cleanColNames(marketing_customers_df.copy(), ncol={"unnamed:_0":"index", "customer_lifetime_value": "lifetime_value", "employmentstatus": "employment_status"})

# Review categorical Data
display(list((marketing_customers_df.select_dtypes(exclude="number")).columns))
# Clean state (as in previous datasets)
marketing_customers_df["state"] = marketing_customers_df["state"].apply(lambda x: abbrev_to_us_state[x] if x in abbrev_to_us_state.keys() else x)
# Review response values
print(list(marketing_customers_df["response"].unique())) # No cleaning needed
# Review coverage values
print(list(marketing_customers_df["coverage"].unique())) # No cleaning needed
# Review education values
print(list(marketing_customers_df["education"].unique())) # No cleaning needed
# Review employment_status values
print(list(marketing_customers_df["employment_status"].unique())) # No cleaning needed
# Review gender values
print(list(marketing_customers_df["gender"].unique())) # No cleaning needed
# Review policy_type values
print(list(marketing_customers_df["policy_type"].unique())) # No cleaning needed
# Review renew_offer_type values
print(list(marketing_customers_df["renew_offer_type"].unique())) # No cleaning needed
# Review sales_channel values
print(list(marketing_customers_df["sales_channel"].unique())) # No cleaning needed
# Review vehicle_class values
print(list(marketing_customers_df["vehicle_class"].unique()))
marketing_customers_df.loc[marketing_customers_df["vehicle_class"].isin(["Sports Car", "Luxury SUV", "Luxury Car"]), "vehicle_class"] = "Luxury" # Normalize with previous datasets
print(list(marketing_customers_df["vehicle_class"].unique()))
# Review vehicle_size values
print(list(marketing_customers_df["vehicle_size"].unique()))
marketing_customers_df.loc[marketing_customers_df["vehicle_size"] == "Medsize", "vehicle_size"] = "Medium" # Rename Medsize as Medium to fit with other two sizes ('Small', 'Large')
print(list(marketing_customers_df["vehicle_size"].unique()))
# Review vehicle_type values
print(list(marketing_customers_df["vehicle_type"].unique())) # No cleaning needed
# Remove $ in lifetime_value
marketing_customers_df["lifetime_value"] = marketing_customers_df["lifetime_value"].apply(lambda x: float(x.replace('$','')) if isinstance(x, str) else x)

# Set Correct DType
marketing_customers_df["lifetime_value"] = marketing_customers_df["lifetime_value"].astype("float64") # As in previous datasets
marketing_customers_df["response"] = marketing_customers_df["response"] == "Yes" # Transform Yes/No string to bool
marketing_customers_df["effective_to_date"] = marketing_customers_df.apply(lambda x: pd.to_datetime(x["effective_to_date"], format="%Y-%m-%d"), axis=1) # To Datetime

# Clean Null Values
display(marketing_customers_df.isna().sum()) # No null values

# Clean Duplicates
cleanDuplicated(marketing_customers_df)

display(marketing_customers_df)
```
